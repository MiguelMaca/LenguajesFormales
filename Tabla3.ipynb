```python
!pip install PyPDF2
import pandas as pd
import re
from PyPDF2 import PdfReader
```

```python
rutaArch = "/content/drive/MyDrive/Analisis Anual 2018 ETAS.pdf"
```

```python
documento = PdfReader(rutaArch)
```

```python
len(documento.pages)
```

```python
pagina = documento.pages[4]
texto = pagina.extract_text()
texto
```

```python
texto.split("\n")
```

```python
patron = "\n(\d+)\s+(\d+)\s+([\w\s\/]+?)\s+([\w\s]+?)\s+([\w\s,-]+?)\s+(\d+)"
```

```python
datos = re.findall(patron,texto)
datos
```

```python
etas = pd.DataFrame(datos, columns=['No.', 'SE', 'Area de Salud', 'ETA Reportada', 'Fuente de Contagio', 'No. de Casos'])
etas
```

```python
etas.dtypes
```

```python
etas['No.'] = etas['No.'].astype('int64')
etas['SE'] = etas['SE'].astype('int64')
etas['Area de Salud'] = etas['Area de Salud'].astype('string')
etas['ETA Reportada'] = etas['ETA Reportada'].astype('string')
etas['Fuente de Contagio'] = etas['Fuente de Contagio'].astype('string')
etas['No. de Casos'] = etas['No. de Casos'].astype('int64')
etas.dtypes
```
