```python
!pip install PyPDF2
import pandas as pd
import re
from PyPDF2 import PdfReader
```

```python
rutaArch = "/content/drive/MyDrive/Analisis Anual 2018 ETAS.pdf"
```

```python
documento = PdfReader(rutaArch)
```

```python
len(documento.pages)
```

```python
pagina = documento.pages[3]
texto = pagina.extract_text()
texto
```

```python
texto.split("\n")
```

```python
patron = r"\n(\d+)\s+(\d+)\s+(\d+)\s+(\d+)\s+(\d+%)\s+(\(\d+/\d+\))"
```

```python
datos = re.findall(patron,texto)
datos
```

```python
etas = pd.DataFrame(datos, columns=['AÑO', 'REPORTADOS', 'INVESTIGADOS', 'INFORMADOS', '% NOTIFICACION DAS','FRACCION'])
etas
```

```python
etas.dtypes
```

```python
etas['AÑO'] = etas['AÑO'].astype('int64')
etas['REPORTADOS'] = etas['REPORTADOS'].astype('int64')
etas['INVESTIGADOS'] = etas['INVESTIGADOS'].astype('int64')
etas['INFORMADOS'] = etas['INFORMADOS'].astype('int64')
etas['% NOTIFICACION DAS'] = etas['% NOTIFICACION DAS'].astype('object')
etas['FRACCION'] = etas['FRACCION'].astype('object')
etas.dtypes
```
